```python
def get_location_info(latitude, longitude):
    """
         .
    
    :param latitude: 
    :param longitude: 
    :return:  
    """
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.reverse((latitude, longitude), exactly_one=True)
    
    if location:
        return location.address
    else:
        return "   ."
```

```python
import re

def extract_coordinates_and_zoom(url):
    """
          URL.
    
    :param url: URL 
    :return:  (, , )
    """
    #          URL
    match = re.search(r"#/([-0-9.]+),([-0-9.]+),(\d+)/", url)
    if match:
        center_lon = float(match.group(1))
        center_lat = float(match.group(2))
        zoom = int(match.group(3))
        return center_lon, center_lat, zoom
    else:
        raise ValueError("       URL")

def calculate_tile_size_in_degrees(zoom):
    """
               .
    
    :param zoom:   
    :return:      
    """
    tiles = 2 ** zoom
    width_degrees = 360.0 / tiles
    height_degrees = 170.1022 / tiles  #      ( -85.0511  85.0511)
    return width_degrees, height_degrees

def calculate_bounding_box(center_lon, center_lat, zoom):
    """
       (    )      .
    
    :param center_lon:   
    :param center_lat:   
    :param zoom:   
    :return:        
    """
    width_degrees, height_degrees = calculate_tile_size_in_degrees(zoom)
    
    #        
    top_left_lat = center_lat + height_degrees / 2
    top_left_lon = center_lon - width_degrees / 2
    bottom_right_lat = center_lat - height_degrees / 2
    bottom_right_lon = center_lon + width_degrees / 2
    
    return (top_left_lon, top_left_lat), (bottom_right_lon, bottom_right_lat)

def main(url):
    try:
        #      URL
        center_lon, center_lat, zoom = extract_coordinates_and_zoom(url)
        
        #   
        top_left, bottom_right = calculate_bounding_box(center_lon, center_lat, zoom)
        
        print(f"  :  {top_left[0]},  {top_left[1]}")
        print(f"  :  {bottom_right[0]},  {bottom_right[1]}")
    except ValueError as e:
        print(f": {e}")

#  
url = "https://map.openaerialmap.org/#/9.058565497398376,48.54593277868612,17/square/1202211003310223222?_k=gdjk3g"
main(url)
```

```python
import os
from PIL import Image

def is_black_image(image_path):
    """,     ."""
    try:
        with Image.open(image_path) as img:
            #     'RGB',     
            img = img.convert('RGB')
            #   
            pixels = img.getdata()
            # ,     (0, 0, 0)
            for pixel in pixels:
                if pixel != (0, 0, 0):
                    return False
            return True
    except Exception as e:
        print(f"    {image_path}: {e}")
        return False

def delete_black_images(folder_path):
    """      ."""
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            try:
                if is_black_image(file_path):
                    os.remove(file_path)
                    print(f"  : {filename}")
            except Exception as e:
                print(f"    {file_path}: {e}")

#       
folder_path = 'black_photo'

delete_black_images(folder_path)
```
