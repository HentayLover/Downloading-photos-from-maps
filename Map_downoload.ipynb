```python
def get_location_info(latitude, longitude):
    """
    Возвращает описание местоположения по заданным координатам.
    
    :param latitude: Широта
    :param longitude: Долгота
    :return: Описание местоположения
    """
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.reverse((latitude, longitude), exactly_one=True)
    
    if location:
        return location.address
    else:
        return "Информация о местоположении недоступна."
```

# Извлекает из квадратов координаты для начальных точек

```python
import re

def extract_coordinates_and_zoom(url):
    """
    Извлекает координаты и уровень зума из URL.
    
    :param url: URL карты
    :return: кортеж (долгота, широта, зум)
    """
    # Регулярное выражение для извлечения координат и уровня зума из URL
    match = re.search(r"#/([-0-9.]+),([-0-9.]+),(\d+)/", url)
    if match:
        center_lon = float(match.group(1))
        center_lat = float(match.group(2))
        zoom = int(match.group(3))
        return center_lon, center_lat, zoom
    else:
        raise ValueError("Не удалось извлечь координаты и зум из URL")

def calculate_tile_size_in_degrees(zoom):
    """
    Вычисляет размер тайла в градусах долготы и широты для заданного уровня зума.
    
    :param zoom: уровень зума карты
    :return: ширина и высота тайла в градусах
    """
    tiles = 2 ** zoom
    width_degrees = 360.0 / tiles
    height_degrees = 170.1022 / tiles  # Средняя высота карты в градусах (от -85.0511 до 85.0511)
    return width_degrees, height_degrees

def calculate_bounding_box(center_lon, center_lat, zoom):
    """
    Вычисляет координаты углов (верхний левый и нижний правый) квадрата в градусах долготы и широты.
    
    :param center_lon: долгота центра карты
    :param center_lat: широта центра карты
    :param zoom: уровень зума карты
    :return: координаты верхнего левого и нижнего правого углов квадрата
    """
    width_degrees, height_degrees = calculate_tile_size_in_degrees(zoom)
    
    # Вычисляем координаты верхнего левого и нижнего правого углов
    top_left_lat = center_lat + height_degrees / 2
    top_left_lon = center_lon - width_degrees / 2
    bottom_right_lat = center_lat - height_degrees / 2
    bottom_right_lon = center_lon + width_degrees / 2
    
    return (top_left_lon, top_left_lat), (bottom_right_lon, bottom_right_lat)

def main(url):
    try:
        # Извлекаем координаты и зум из URL
        center_lon, center_lat, zoom = extract_coordinates_and_zoom(url)
        
        # Вычисляем координаты углов
        top_left, bottom_right = calculate_bounding_box(center_lon, center_lat, zoom)
        
        print(f"Верхний левый угол: Долгота {top_left[0]}, Широта {top_left[1]}")
        print(f"Нижний правый угол: Долгота {bottom_right[0]}, Широта {bottom_right[1]}")
    except ValueError as e:
        print(f"Ошибка: {e}")

# Пример использования
url = "https://map.openaerialmap.org/#/9.058565497398376,48.54593277868612,17/square/1202211003310223222?_k=gdjk3g"
main(url)
```

# Удаление черных фото

```python
import os
from PIL import Image

def is_black_image(image_path):
    """Проверяет, является ли изображение полностью черным."""
    try:
        with Image.open(image_path) as img:
            # Преобразуем изображение в режим 'RGB', если оно в другом формате
            img = img.convert('RGB')
            # Получаем данные изображения
            pixels = img.getdata()
            # Проверяем, что все пиксели черные (0, 0, 0)
            for pixel in pixels:
                if pixel != (0, 0, 0):
                    return False
            return True
    except Exception as e:
        print(f"Не удалось обработать файл {image_path}: {e}")
        return False

def delete_black_images(folder_path):
    """Удаляет все черные изображения в указанной папке."""
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            try:
                if is_black_image(file_path):
                    os.remove(file_path)
                    print(f"Удалено черное изображение: {filename}")
            except Exception as e:
                print(f"Ошибка при удалении файла {file_path}: {e}")

# Укажите путь к вашей папке с изображениями
folder_path = 'black_photo'

delete_black_images(folder_path)
```
